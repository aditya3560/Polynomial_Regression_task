```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data={"Temperature_c":[20,22,24,26,28,30,32,34,36],"Ice_Cream_Sales":[120,140,175,220,280,340,390,420,430]}
```

```python
data=pd.DataFrame(data)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.info()
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.scatterplot(x="Temperature_c",y="Ice_Cream_Sales",data=data)
```

```python
x=data[["Temperature_c"]].values
y=data[["Ice_Cream_Sales"]].values
```

```python
x
```

```python
y
```

```python
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
lin_reg.score(x,y)
```

```python
pred1=lin_reg.predict(x)

plt.figure(figsize=(12,6),dpi=200)
plt.scatter(x,y,c='red',label='Actual Data')
plt.plot(x,pred1,c='black',label='Regression Line')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
x_poly3 =poly.fit_transform(x)
x_poly3
```

```python
poly_reg3=LinearRegression()
poly_reg3.fit(x_poly3,y)

poly_reg3.score(x_poly3,y)
```

```python
pred3=poly_reg3.predict(x_poly3)
plt.figure(figsize=(12,6),dpi=200)
plt.scatter(x,y,c='red',label='Actual Data')
plt.plot(x,pred3,c='black',label='Regression Curve')
plt.legend()
plt.show()
```
